#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def func1(stroka):
    c=0
    letters={'a','e','i','o','u'}
    for x in stroka.lower():
        if x in letters:
            c+=1
        if ord(x) not in range (97,123):
            return "В строке содержатся недопустимые символы"
    return c
assert func1("aeiou") == 5, "Ошибка: не посчитал все гласные"
assert func1("hmn") == 0, "Ошибка: нашел гласные там, где их нет"
assert func1("") == 0, "Ошибка: не сработал на пустой строке"
assert func1("AbcDe") == 2, "Ошибка: не учел заглавные буквы"
assert func1("123! @") == "В строке содержатся недопустимые символы", "Ошибка: не сработал на спецсимволах"
assert func1("асбca") == "В строке содержатся недопустимые символы", "Ошибка: посчитал кириллицу"
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def strcount (stroka):
    d=dict()
    for x in stroka:
        if x in d:
            return False
        else:
            d[x]=0
    return True
assert strcount("aeiou") == True, "Ошибка"
assert strcount("hmm") == False, "Ошибка"
assert strcount("hmmdsdsfsgsdf0r39429!!") == False, "Ошибка"
assert strcount("hmfsgd0r!?") == True, "Ошибка"
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def func3(a):
    if a<=0:
        return 'Число должно быть положительным'
    b=str(bin(a)).count('1')

    return b
assert func3(1) == 1, "Ошибка"
assert func3(2) == 1, "Ошибка"
assert func3(3) == 2, "Ошибка"
assert func3(4) == 1, "Ошибка"
assert func3(0) == 'Число должно быть положительным', 'Ошибка: Число должно быть положительным'
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
from functools import reduce
def magic(n, c):
    if n<10:
        return c
    s=str(n)
    l=[int(x) for x in s]
    n=reduce(lambda a,b:a*b, l)
    c+=1
    return magic(n, c)
c=0
assert magic(39, c) == 3, 'Ошибка'
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(pred, true):
    if not pred or len(pred)!=len(true):
        return "Векторы должны быть одинаковой длины!"
    sko=sum((p-t)**2 for p,t in zip(pred, true))/len(pred)
    return sko
assert mse([1, 2, 3], [1, 2, 3]) == 0.0
assert round(mse([1, 2, 3], [1, 0, 5]), 3) == 2.667
assert mse([-1], [1]) == 4.0
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def get_min_primes(limit):
    """Строит массив минимальных простых делителей до limit."""
    min_prime = list(range(limit + 1))
    for i in range(2, int(limit**0.5) + 1):
        if min_prime[i] == i:  # Если i простое
            for j in range(i*i, limit + 1, i):
                if min_prime[j] == j:
                    min_prime[j] = i
    return min_prime

def prime_factors_sieve(n):
    if n <= 1: return ""
    
    # 1. Готовим решето (в реальной задаче лучше делать это один раз вне функции)
    min_prime = get_min_primes(n)
    
    # 2. Быстрое разложение
    factors = {}
    temp = n
    while temp > 1:
        p = min_prime[temp]
        factors[p] = factors.get(p, 0) + 1
        temp //= p
        
    # 3. Форматирование строки
    res = ""
    # Сортируем ключи, чтобы множители шли по возрастанию
    for p in sorted(factors.keys()):
        count = factors[p]
        if count == 1:
            res += f"({p})"
        else:
            res += f"({p}**{count})"
    return res

# --- ТЕСТЫ ---
assert prime_factors_sieve(86240) == "(2**5)(5)(7**2)(11)", "Ошибка на 86240"
assert prime_factors_sieve(10) == "(2)(5)", "Ошибка на 10"
assert prime_factors_sieve(2) == "(2)", "Ошибка на 2"

print("Все тесты пройдены!")
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(n):
    s=0
    for x in range(1,n):
        s=s+x**2
        if s==n:
            return x
        if s>n:
            return 'It is impossible'
assert pyramid(5)==2, "Ошибка"
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def func8(a):
    s=str(a)
    l=[int(x) for x in s]
    lenl=len(l)//2
    if len(l)%2==1:
        if sum(l[:lenl])==sum(l[lenl+1:]):
            return True
    else:
        if sum(l[:lenl-1])==sum(l[lenl+1:]):
            return True
    return False

assert func8(12321)== True, "Ошибка"
assert func8(123321)== True, "Ошибка"
assert func8(12311)== False, "Ошибка"
```
